```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_img(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img,channels=3)
    img = tf.image.convert_image_dtype(img,tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
content_image = load_img('xp.jpg')
style_image = load_img('reference.jpg')
```

```python
plt.imshow(np.squeeze(content_image))
plt.show()
```

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
stylized_images = model(tf.constant(content_image),tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(stylized_images))
plt.show()
```

```python
cv2.imwrite('generated_img.jpg',cv2.cvtColor(np.squeeze(stylized_images)*255, cv2.COLOR_BGR2RGB))
```
